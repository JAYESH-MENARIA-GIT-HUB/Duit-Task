### Task 1: Object Detection

```python
# Importing libaries
import cv2
import matplotlib.pyplot as plt
```

```python
# Threshold to detect object
thres = 0.56
```

```python
#Labels
classNames= []
classFile = r'coco.names'
with open(classFile,'rt') as f:
    classNames = f.read().rstrip('\n').split('\n')
```

```python
#loading pre-trained model from disk
configPath = r'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath = r'frozen_inference_graph.pb'
```

```python
#defining model
net = cv2.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/ 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)
```

### Testing model on Image

```python
#loading image
image=r'check.jpg'
img = cv2.imread(image)
```

```python
classIds, confs, bbox = net.detect(img,confThreshold=thres)
if len(classIds) != 0:
    for classId, confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
        # draw the bounding box on the image
        cv2.rectangle(img,box,color=(255,0,0),thickness=3)
        cv2.putText(img,classNames[classId-1].upper(),(box[0]+5,box[1]+30),
                    cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
        cv2.putText(img,str(round(confidence*100,2)),(box[0]+5,box[1]+55),
                    cv2.FONT_HERSHEY_COMPLEX,0.6,(255,0,0),2)
```

```python
#show result
plt.imshow(img)
```

### Testing model on Video

```python
# a function for timestamp for video
def frame_time(cap,curr_frame,prev,minute):
    current = round(cap.get(cv2.CAP_PROP_POS_MSEC)/1000,2)
    if current > 60:
        new = round(current - prev,2)
        cv2.putText(curr_frame,str(minute+1)+"."+str(new),(30,100),
                    cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),1)
        if int(new)%60==0 and int(new)>0:
            prev = round(cap.get(cv2.CAP_PROP_POS_MSEC)/1000,2)
            minute += 1 
    else:
        cv2.putText(curr_frame,str(current),(30,100),
                    cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),1)
        prev = round(cap.get(cv2.CAP_PROP_POS_MSEC)/1000,2)
        
    return curr_frame,prev,minute
```

```python
#loading video
cap = cv2.VideoCapture("videos/Honda_Civic.mov")

in_one_frame = []
asec_frame = {}
prev,minute = 0,0
total_frame = int(cap.get(cv2.CAP_PROP_FPS))
sec = 0

current_frame = 0
while True:
    success,img = cap.read()
    if success==True:
        classIds, confs, bbox = net.detect(img,confThreshold=.60)
        #print(classIds,bbox)
        img,prev,minute = frame_time(cap,img,prev,minute)
        if len(classIds) != 0:
            for classId, confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
                cv2.rectangle(img,box,(0,0,255),1)
                cv2.putText(img,classNames[classId-1].upper(),(box[0]+10,box[1]+30),
                            cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,255),1)
                in_one_frame.append(classNames[classId-1])


        cv2.imshow("Output Video",img)
        if current_frame%total_frame==0:
            asec_frame[sec] = {i:in_one_frame.count(i) for i in in_one_frame}
            in_one_frame = []
            sec += 1
        current_frame += 1
        if cv2.waitKey(1) & 0xFF ==ord('q'):
            cv2.destroyAllWindows()
            break
    else:
        break
#close all opencv windows open
cap.release()
cv2.destroyAllWindows()
```

## categories appearing in micro-batch segment

```python
# get class in range between secondes
def det_item(start,stop,listofsec,only_labels=False):
    a = {}
    for i in range(start,stop+1):
        for j,k in zip(listofsec[i].keys(),listofsec[i].values()):
            if j in a:
                a[j] = a[j] + k
            else:
                a[j] = k
    if only_labels:
        return a.keys()
    return a
```

```python
start = 1 #in second 
stop =  7 #in second
listofsec = asec_frame
det_item(start,stop,listofsec,only_labels=False) # car and person are detected in 1 to 7 second 
```
